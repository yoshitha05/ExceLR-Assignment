```python
import nltk
from nltk.tokenize import word_tokenize

# Corrected line
nltk.download('all')  # Downloads all NLTK data

# Example usage of word_tokenize
text = "Hello, how are you doing today?"
tokens = word_tokenize(text)
print("Tokens:", tokens)
```

***Tokenization***

```python
text = "Hello, world! This is sample text for tokenization."
tokens = word_tokenize(text)
print("Tokens:", tokens)
```

***Stemming***

```python
import nltk
from nltk.stem import PorterStemmer
words = ["running", "happier", "better", "dogs", "jumping", "easily"]
stammer = PorterStemmer()
stemmed_words = [stammer.stem(word) for word in words]
print("Stemmed Words:", stemmed_words)
```

***Lemmatization***

```python
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
words = ["running", "better", "cats", "jumping", "easily", "went"]
lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word, pos='v') for word in words]
print("Lemmatized Words:", lemmatized_words)
```
